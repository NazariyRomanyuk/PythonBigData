1. (6б). Створіть клас Building, додайте три атрибути, один з яких має значення за замовчуванням та два методи на свій розсуд.

```python
class Building:
  def __init__(self, height, story_count, is_concrete=True):
      self.height = height
      self.story_count = story_count
      self.is_concrete = is_concrete
  def __str__(self):
      res = f"This {self.story_count} floor building is {self.height}m tall and"
      if (self.is_concrete):
        res += " is concrete."
      else:
        res += " is not concrete."
      return res
  def calculate_story_height(self):
      if self.story_count == 0:
        return 0
      return self.height/self.story_count
```

5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class ThreeDShape:
  def get_volume(self):
    return 0
  def __add__(self, other):
    return self.get_volume() + other.get_volume()
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math
class Cylinder(ThreeDShape):
  def __init__(self, base_radius, height):
    self.base_radius = base_radius
    self.height = height

  def get_volume(self):
    return math.pi * self.base_radius**2 * self.height

class Cube(ThreeDShape):
  def __init__(self, face_length):
    self.face_length = face_length

  def get_volume(self):
    return self.face_length ** 3

cylinder = Cylinder(5, 10)
cube = Cube(5)
print(cylinder + cube)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  __next_free = 1
  def __init__(self, balance=0):
    self.balance = balance
    self.__id = EWallet.__next_free
    EWallet.__next_free += 1
  @property
  def balance(self):
    return self.__balance

  @balance.setter
  def balance(self, value):
    if value < 0:
      print("Balance cannot be negative! Automatically set to 0.")
      self.__balance = 0
    else:
      self.__balance = value

  @property
  def id(self):
    return self.__id

  def top_up(self, value):
    self.balance += value

  def withdraw(self, value):
    self.balance -= value


wallet = EWallet(100)
print(f"Created wallet with balance: {wallet.balance}")
wallet.top_up(200)
print(f"Topped up wallet with 200: {wallet.balance}")
wallet.withdraw(10)
print(f"Withdrew 10 from wallet: {wallet.balance}")
wallet.withdraw(500)
print(f"Withdrew 500 from wallet (underflow): {wallet.balance}")
print(f"wallet has an id of {wallet.id}")
wallet1 = EWallet()
print(f"Crated wallet1 with an id of {wallet1.id}")

```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self, name, price):
    self.name = name
    self.price = price

  @property
  def name(self):
    return self._name

  @name.setter
  def name(self, value):
    self._name = value

  @property
  def price(self):
    return self._price

  @price.setter
  def price(self, value):
    if (value < 0):
      print("Price cannot be negative! Automatically set to 0.")
      self._price = 0
    else:
      self._price = value

  def __str__(self):
    return f"{self.name.capitalize()}: ${self.price:.2f}"

class ShoppingCart:

  def __init__(self, *args):
    self._items = [x for x in args]

  def add_item(self, item):
    if not isinstance(item, Item):
      print("Can only add Items to Cart!")
    else:
      self._items.append(item)

  def remove_item(self, item_name):
    self._items = [x for x in self._items if x.name != item_name]

  def __str__(self):
    return str([str(x) for x in self._items])

cake = Item("cake", 14.99)
shampoo = Item("shampoo", 4.99)
lightbulb = Item("lightbulb", 9.99)
glitched_item = Item("glitch", -5)

cart = ShoppingCart(cake, shampoo, lightbulb)
print(cart)

cart.add_item(Item("toothpaste", 2.99))
print(cart)

cart.remove_item("shampoo")
print(cart)
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class DayOfTheWeekIterator:
  def __init__(self):
    self.__days = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")
    self.__index = 0
  def __iter__(self):
    return self
  def __next__(self):
    if self.__index < len(self.__days):
      res = self.__days[self.__index]
      self.__index += 1
      return res
    else:
      raise StopIteration
dotw_iter = DayOfTheWeekIterator()
for day in dotw_iter:
  print(day)
```
