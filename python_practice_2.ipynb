# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
my_set = set()
my_set.update([1,2,2,5,6,7])
my_set
```

Так, множини є змінним типом даних. До одного й того ж об'єкта множини в пам'яті можна на місці додавати та видаляти елементи. Проте елементи множини мають бути незмінними, адже змінні типи даних не хешуються.

18. Створити множину, маючи список my_list = [1, 1, 2, 67, 67, 8, 9]. Пояснити, чому "зникли" деякі елементи.

```python
my_list = [1, 1, 2, 67, 67, 8, 9]
my_set = set(my_list)
my_set
```

"Зникнення" деяких елементів є виявленням основної властивості множин - неповторність їх елементів. Саме тому легко "позбуватися" дублікатів у інших колекціях таким чином, отримуючи лише унікальні елементи при створення множини, використовуючи колекцію як параметр конструктора.

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
my_set_1 = {1,2,2,5,6,8,8,8,9}
my_set_2 = {1,3,5,5,8,10}
union_set = my_set_1.union(my_set_2)
difference_set = my_set_1.difference(my_set_2)
intersection_set = my_set_1.intersection(my_set_2)
symm_diff_set = my_set_1.symmetric_difference(my_set_2)
(f"Union: {union_set}, Difference (1-2): {difference_set}, Intersection: {intersection_set}, Symmetric Difference: {symm_diff_set}")
```
